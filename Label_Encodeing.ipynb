```python
import pandas as pd # Ensure pandas is available
from sklearn.preprocessing import LabelEncoder

# Re-read the dataset to ensure 'dataset' is a DataFrame
dataset = pd.read_excel("/content/ml_practice_dataset 2.xlsx")

le = LabelEncoder()
# Convert the 'Name' column to string type and fill NaN values before encoding
name_series = dataset["Name"].astype(str).fillna('')
le.fit(name_series) # Fit the encoder
dataset["Name"] = le.transform(name_series) # Transform the data
dataset
```
